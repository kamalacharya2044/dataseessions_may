```python
import numpy as np
mdf1["HitPoint"] = np.NaN
```

```python
mdf1["HitPoint"] = mdf1["Defense"]/ mdf1["Attack"]
```

```python
mdf1
```

```python
mdf1["Defense"]/ mdf1["Attack"]
```

```python
mdf1["Defense"].max()
```

```python
mdf1["NormalizeDefense"] = mdf1["Defense"]/ mdf1["Defense"].max()
```

```python
mdf1
```

```python
mdf1["NormalizeAttack"] = mdf1["Attack"]/ mdf1["Attack"].max()
```

```python
mdf1["NormalizeSpeed"] = mdf1["Speed"]/ mdf1["Speed"].max()
```

```python
mdf1
```

```python
ax = mdf1.plot(kind="bar" , y="NormalizeAttack")
mdf1.plot(kind="bar", y="NormalizeDefense", ax=ax, color="C2")
mdf1.plot(kind="bar", y="NormalizeSpeed", ax=ax, color="C3")
```
